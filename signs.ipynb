```python
import csv
import pandas as pd
import numpy as np

df1 = pd.read_csv("E:/elevatedjugularvenouspressure.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df2 = pd.read_csv("E:/hepatojugularreflux.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df3 = pd.read_csv("E:/galloprhythm.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df4 = pd.read_csv("E:/displacedapicalimpulse.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df5 = pd.read_csv("E:/cardiacmurmur.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df6 = pd.read_csv("E:/peripheraloedema.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df7 = pd.read_csv("E:/pulmonarycrepitation.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df8 = pd.read_csv("E:/pleuraleffusion.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df9 = pd.read_csv("E:/tachycardia.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df10 = pd.read_csv("E:/irregularpulse.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df11 = pd.read_csv("E:/tachypnoea.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df12 = pd.read_csv("E:/hepatomegaly.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df13 = pd.read_csv("E:/cachexia.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
df14 = pd.read_csv("E:/ascites.csv", 
                  error_bad_lines=False,sep=';', encoding = 'utf-8')
```

```python
dfa = df1.merge(df2, on=["subject_id"], how='outer')
```

```python
dfb = dfa.merge(df3, on=["subject_id"], how='outer')
```

```python
dfc = dfb.merge(df4, on=["subject_id"], how='outer')
```

```python
dfd = dfc.merge(df5, on=["subject_id"], how='outer')
```

```python
dfe = dfd.merge(df6, on=["subject_id"], how='outer')
```

```python
dff = dfe.merge(df7, on=["subject_id"], how='outer')
```

```python
dfg = dff.merge(df8, on=["subject_id"], how='outer')
```

```python
dfh = dfg.merge(df9, on=["subject_id"], how='outer')
```

```python
dfi = dfh.merge(df10, on=["subject_id"], how='outer')
```

```python
dfj = dfi.merge(df11, on=["subject_id"], how='outer')
```

```python
dfk = dfj.merge(df12, on=["subject_id"], how='outer')
```

```python
dfl = dfk.merge(df13, on=["subject_id"], how='outer')
```

```python
dffinal = dfl.merge(df14, on=["subject_id"], how='outer')
```

```python
dffinal.to_csv("E:/signs.csv",index=False)
```
